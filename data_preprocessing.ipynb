```r
df_original <- read.csv("working_population_health_data.csv", header=T, sep=",")
# df <- subset(df, subset <- (v4 != 98) & (v7 != 98) & (v8 != 98) & (v24 != 99) &
#               !(v41 %in% c(998, 999)) & !(v42 %in% c(98, 99)) & !(v43 %in% c(98, 99)) &
#               !(v33 %in% c(97, 98, 99)) & (v34 != 98))
df <- df_original
head(df)

# 定義變數意義
ageKey <- c("2"="15-19歲", "3"="20-29歲", "4"="30-39歲", "5"="40-49歲", "6"="50-59歲", "7"="60-69歲", "8"="60-69歲")
df$v4 <- ageKey[as.character(df$v4)]
df$v4_original <- df_original$v4

genderKey <- c("1"="男性", "2"="女性")
df$v45 <- genderKey[as.character(df$v45)]
df$v45_original <- df_original$v45

# 0: 否, 1: 是
yesNoKey <- c("1"="1", "2"="0")
df$v43 <- yesNoKey[as.character(df$v43)]
df$v24 <- yesNoKey[as.character(df$v24)]
head(df)

# 1: 完全不可能 4: 非常可能
aiKey <- c("1"="4", "2"="3", "3"="2", "4"="1")
df$v42 <- aiKey[as.character(df$v42)]
```

```r
# Age
ageDist <- data.frame(table(Age = df$v4))
ageDist
ageGraph <- ggplot(ageDist, aes(x=Age, y=Freq)) + geom_col()
ageGraph
```

```r
# Gender
genderDist <- data.frame(table(Gender = df$v45))
genderDist
genderGraph <- ggplot(genderDist, aes(x=Gender, y=Freq)) + geom_col()
genderGraph
```

```r
# Remote work
remoteDist <- data.frame(table(df$v24))
names(remoteDist) <- c("Remote_3months", "Frequency")
remoteDist$Remote_3months <- c("Yes", "No")
remoteDist
remoteGraph <- ggplot(remoteDist, aes(x=Remote_3months, y=Frequency)) + geom_col()
remoteGraph
```
